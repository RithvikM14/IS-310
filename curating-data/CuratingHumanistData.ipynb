```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv
```

```python
base_url='https://humanist.kdl.kcl.ac.uk/Archives/'
base_response = requests.get(base_url)
base_soup = BeautifulSoup(base_response.text, features="html.parser")


main_links = base_soup.find_all('a')[1:]
main_links
```

```python
convertedbase_url = base_url + main_links[0].get('href')
sub_base_response = requests.get(convertedbase_url)
sub_base_soup = BeautifulSoup(sub_base_response.text, "html.parser")
archive_links = sub_base_soup.find_all('a')[1:-1]
archive_links 
```

```python
archive_links[0].get_text().split(".")[1]
```

```python
with open('archive_data.csv', 'w', newline='',encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Link', 'Text','Years'])
    for link in archive_links:
        volume_base_url = convertedbase_url + link.get('href')
        volume_base_response = requests.get(volume_base_url)
        volume_base_soup = BeautifulSoup(volume_base_response.text, "html.parser")
        writer.writerow([link.get_text(), volume_base_soup.get_text()[:10000],link.get_text().split(".")[1]])

archive_data = pd.read_csv('archive_data.csv')

```

```python
currentbase_url = base_url + main_links[1].get('href')
sub_base_response = requests.get(currentbase_url)
sub_base_soup = BeautifulSoup(sub_base_response.text, "html.parser")
current_links = sub_base_soup.find_all('a')[1:]
current_links 
```

```python
with open('current_data.csv', 'w', newline='',encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Link', 'Text','Volume'])
    for link in current_links:
        volume_base_url = currentbase_url + link.get('href')
        volume_base_response = requests.get(volume_base_url)
        volume_base_soup = BeautifulSoup(volume_base_response.text, "html.parser")
        writer.writerow([link.get_text(), volume_base_soup.get_text()[:10000],link.get_text().split(".")[1]])

current_data = pd.read_csv('current_data.csv')
current_data
```

```python
concatenated_data = pd.concat([archive_data, current_data], ignore_index=True, sort=False)
concatenated_data
```

```python

concatenated_data.to_csv('web_scraped_humanist_listserv.csv', index=False)
concatenated_data = pd.read_csv('web_scraped_humanist_listserv.csv')
concatenated_data
```
