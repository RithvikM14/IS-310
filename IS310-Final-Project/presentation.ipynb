## Literature Review

The use of Twitter among different cultural groups and the global influence of the Indian Premier League (IPL) offer a unique perspective on how technology and sports intersect in modern society. André Brock (2012) explores how Twitter has become a significant platform for Black Americans, serving not just for communication but as a space where cultural identities are expressed and shaped. This idea helps us understand Twitter as more than a tool—it's a place where minority voices can perform and share their cultural experiences. Colin Agur (2013) views the IPL not just as a cricket league, but as an event that has reshaped the business and culture of cricket worldwide. He discusses how the IPL has changed traditional cricket culture, introducing it as a major spectacle that engages young fans and changes global sports practices. His insights help frame the IPL as a cultural phenomenon that affects how people see and interact with the sport on platforms like Twitter. By combining these perspectives, this project investigates how the 2021 IPL is discussed on Twitter, focusing on the cultural conversations it triggers among its global audience. Analyzing Twitter data through the frameworks provided by Brock and Agur will reveal how IPL influences and reflects broader cultural and technological trends, enhancing our understanding of how global sports events and digital platforms interact in shaping cultural discussions.


Brock, A. (2012). From the Blackhand Side: Twitter as a Cultural Conversation. Journal of Broadcasting & Electronic Media, 56(4), 529–549. https://doi.org/10.1080/08838151.2012.732147
