# Cultural Analysis of IPL 2021 Through Twitter Data

## Introduction

This project dives into the emotional and social currents surrounding the Indian Premier League (IPL) 2021, as reflected on Twitter. We explore not only the general sentiment towards the event but also the ways in which social networks mediate and influence the spread of these emotions.

## Data Collection & Curation

Data was collected from Twitter using the `#ipl2021` hashtag during the event. This dataset was then curated to facilitate a deeper analysis of the emotional and social dimensions of the public discourse.

*Figure 1: Tweet Volume Over Time*

**Analysis**: The graph represents the ebbs and flows of collective focus. ### Tweet Sentiment Distribution

![Tweet Sentiment Distribution](viz-2.png)

**Analysis**: The distribution of sentiments sheds light on the emotional responses elicited by IPL 2021. With a significant share of tweets being neutral, the platform served as a space for discussion and commentary. Positive sentiments paint a picture of celebration and camaraderie, while negative sentiments likely point to moments of controversy or defeat, each emotion spreading through social networks, influencing and being influenced by the community's mood.

## Conclusion & Reflection

Through the lens of Twitter, the cultural analysis of IPL 2021 highlights how individual emotions are expressed and shared, contributing to a collective emotional narrative. Social networks not only reflect these emotions but also act as channels through which emotional reactions are amplified and spread. This analysis provides insight into the relationship between individual emotional expressions and their social amplification during shared cultural events.
